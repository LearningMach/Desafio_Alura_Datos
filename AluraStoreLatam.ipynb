### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
#En este primer análisis, debes calcular el ingreso total de cada tienda.
#Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

precio_tienda = sum(tienda['Precio'])
precio_tienda2 = sum(tienda2['Precio'])
precio_tienda3 = sum(tienda3['Precio'])
precio_tienda4 = sum(tienda4['Precio'])

print(f"La tienda tuvo un ingreso total de $ {precio_tienda}")
print(f"La tienda 2 tuvo un ingreso total de $ {precio_tienda2}")
print(f"La tienda 3 tuvo un ingreso total de $ {precio_tienda3}")
print(f"La tienda 4 tuvo un ingreso total de $ {precio_tienda4}")
```

```python
import matplotlib.pyplot as plt

nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [precio_tienda, precio_tienda2, precio_tienda3, precio_tienda4]

plt.barh(nombres_tiendas, ingresos, color='cornflowerblue')
plt.xlabel('Ingreso Total ($)')
plt.title('Ingresos Totales por Tienda')
plt.tight_layout()
plt.show()

print()
plt.pie(ingresos, labels=nombres_tiendas, autopct='%1.2f%%', startangle=90, colors=['lightcoral', 'gold', 'lightgreen', 'skyblue'])
plt.title('Distribución de Ingresos Totales por Tienda')
plt.axis('equal')
plt.show()

print()
plt.plot(nombres_tiendas, ingresos, marker='o', color='b', linestyle='-', linewidth=2, markersize=8)
plt.title('Ingresos Totales por Tienda')
plt.xlabel('Tiendas')
plt.ylabel('Ingreso Total ($)')
plt.grid(True)
plt.show()
```

# 2. Ventas por categoría

```python
#En este debes calcular la cantidad de productos vendidos por categoría en cada tienda.
#La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

"""#Tienda 1
ventas = tienda['Categoría del Producto'].value_counts()
producto_mas_vendido = ventas.idxmax()
veces_vendido = ventas.max()
#Tienda 2
ventas2 = tienda2['Categoría del Producto'].value_counts()
producto_mas_vendido = ventas2.idxmax()
veces_vendido = ventas2.max()
#Tienda 3
ventas3 = tienda3['Categoría del Producto'].value_counts()
producto_mas_vendido = ventas3.idxmax()
veces_vendido = ventas3.max()
# Tienda 4
ventas4 = tienda4['Categoría del Producto'].value_counts()
producto_mas_vendido = ventas4.idxmax()
veces_vendido = ventas4.max()
print(ventas)
print(producto_mas_vendido)
print(veces_vendido)"""

def ventas_por_categoria(tiendas):
  ventas = tiendas['Categoría del Producto'].value_counts()
  producto_mas_vendido = ventas.idxmax()
  veces_vendido = ventas.max()
  return producto_mas_vendido, veces_vendido
#tienda 1
producto1, veces1 = ventas_por_categoria(tienda)
print(f'En la tienda 1 el producto mas vendido fue {producto1} con {veces1} ventas')
#tienda 2
producto2, veces2 = ventas_por_categoria(tienda2)
print(f'En la tienda 2 el producto mas vendido fue {producto2} con {veces2} ventas')
#tienda 3
producto3, veces3 = ventas_por_categoria(tienda3)
print(f'En la tienda 3 el producto mas vendido fue {producto3} con {veces3} ventas')
#tienda 4
producto4, veces4 = ventas_por_categoria(tienda4)
print(f'En la tienda 4 el producto mas vendido fue {producto4} con {veces4} ventas')
```

```python
import matplotlib.pyplot as plt

nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
producto_mas_vendido = [producto1, producto2, producto3, producto4]
veces_vendida = [veces1, veces2, veces3, veces4]
etiqueta = [f'{tienda}\n({producto})' for tienda, producto in zip(nombres_tiendas, producto_mas_vendido)]

plt.barh(etiqueta, veces_vendida, color='orange')
plt.title('Ventas por Categoria')
plt.xlabel('Ventas Totales ($)')
plt.ylabel('')
plt.tight_layout()
plt.show()
print()
plt.pie(veces_vendida, labels=etiqueta, autopct='%1.2f%%', startangle=90, colors=['lightcoral', 'gold', 'lightgreen', 'skyblue'])
plt.title('Distribución de Ventas por Categoria')
plt.axis('equal')
plt.show()
print()
plt.plot(etiqueta, veces_vendida, marker='o', color='b', linestyle='-', linewidth=2, markersize=8)
plt.title('Ventas totales por Categoria')
plt.xlabel('Tiendas')
plt.ylabel('Ingreso Total ($)')
plt.grid(True)
plt.show()

```

# 3. Calificación promedio de la tienda


```python
#En este paso, debes calcular las calificaciones promedio de los clientes para cada tienda.
#El objetivo es conocer la satisfacción del cliente con los productos vendidos.
def definir_promedio(promedios):
    clasificacion = promedios['Calificación']
    suma = sum(clasificacion)
    promedio = suma / len(clasificacion)
    estado = ''
    color = ''
    if 1 <= promedio <= 2:
        estado = 'Muy insatisfecho'
        color = 'red'
    elif 2 < promedio <= 3:
        estado = 'Insatisfecho'
        color = 'orange'
    elif 3 < promedio <= 4:
        estado = 'Satisfecho'
        color = 'skyblue'
    elif 4 < promedio <= 5:
        estado = 'Muy satisfecho'
        color = 'green'
    return promedio, estado, color


#tienda1
promedio1, estado1, color1 = definir_promedio(tienda)
print(f'La tienda 1 tuvo un promedio de {promedio1:.2f} estando el cliente {estado1}')
#tienda2
promedio2, estado2, color2 = definir_promedio(tienda2)
print(f'La tienda 2 tuvo un promedio de {promedio2:.2f} estando el cliente {estado2}')
#tienda3
promedio3, estado3, color3 = definir_promedio(tienda3)
print(f'La tienda 3 tuvo un promedio de {promedio3:.2f} estando el cliente {estado3}')
#tienda4
promedio4, estado4, color4 = definir_promedio(tienda4)
print(f'La tienda 4 tuvo un promedio de {promedio4:.2f} estando el cliente {estado4}')


```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
promedios = [promedio1, promedio2, promedio3, promedio4]
colores = [color1, color2, color3, color4]

fig, ax = plt.subplots()

ax.barh(nombres_tiendas, promedios, color=colores)
ax.set_title('Clasificacion promedio por Tienda')
ax.set_xlabel('Promedio')
ax.set_ylabel('Tiendas')

# Leyenda
leyenda = [
    mpatches.Patch(color='green', label='Muy satisfecho'),
    mpatches.Patch(color='skyblue', label='Satisfecho'),
    mpatches.Patch(color='orange', label='Insatisfecho'),
    mpatches.Patch(color='red', label='Muy insatisfecho')
]

ax.legend(handles=leyenda, loc='upper left', bbox_to_anchor=(1.05, 1))
plt.subplots_adjust(right=0.75)
plt.show()

print()
plt.pie(promedios, labels=nombres_tiendas, autopct='%1.2f%%', startangle=90, colors=['lightcoral', 'gold', 'lightgreen', 'skyblue'])
plt.title('Promedio por Tienda')
plt.axis('equal')
plt.show()

print()
plt.plot(nombres_tiendas, promedios, marker='o', color='b', linestyle='-', linewidth=2, markersize=8)
plt.title('Promedio de ventas por tienda')
plt.xlabel('Tiendas')
plt.ylabel('Promedios')
plt.grid(True)
plt.show()
```

# 4. Productos más y menos vendidos

```python
#En este paso, debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda.
#Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.

def productos(productos):
  ventas = productos['Categoría del Producto'].value_counts()
  producto_mas_vendido = ventas.idxmax()
  veces_vendido = ventas.max()
  producto_menos_vendido = ventas.idxmin()
  veces_menos_vendido = ventas.min()
  return producto_mas_vendido, veces_vendido, producto_menos_vendido, veces_menos_vendido

#tienda1
producto1, veces1, producto2, veces2 = productos(tienda)
print(f'En la tienda 1 el producto mas vendido fue {producto1} con {veces1} ventas y el menos vendido fue {producto2} con {veces2} ventas')
#tienda2
producto3, veces3, producto4, veces4 = productos(tienda2)
print(f'En la tienda 2 el producto mas vendido fue {producto3} con {veces3} ventas y el menos vendido fue {producto4} con {veces4} ventas')
#tienda3
producto5, veces5, producto6, veces6 = productos(tienda3)
print(f'En la tienda 3 el producto mas vendido fue {producto5} con {veces5} ventas y el menos vendido fue {producto6} con {veces6} ventas')
#tienda4
producto7, veces7, producto8, veces8 = productos(tienda4)
print(f'En la tienda 4 el producto mas vendido fue {producto7} con {veces7} ventas y el menos vendido fue {producto8} con {veces8} ventas')

```

```python
import matplotlib.pyplot as plt
import numpy as np

nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
producto_mas_vendido = [producto1, producto3, producto5, producto7]
producto_menos_vendido = [producto2, producto4, producto6, producto8]
veces_mas_vendido = [veces1, veces3, veces5, veces7]
veces_menos_vendido = [veces2, veces4, veces6, veces8]

y = np.arange(len(nombres_tiendas))
ancho = 0.35

plt.barh(y - ancho/2, veces_mas_vendido, height=ancho, color='green', label='Más vendido')
plt.barh(y + ancho/2, veces_menos_vendido, height=ancho, color='red', label='Menos vendido')

for i in range(len(nombres_tiendas)):
    plt.text(veces_mas_vendido[i] + 1, y[i] - ancho/2, producto_mas_vendido[i], va='center', fontsize=8)
    plt.text(veces_menos_vendido[i] + 1, y[i] + ancho/2, producto_menos_vendido[i], va='center', fontsize=8)

plt.xlabel('Cantidad de ventas')
plt.yticks(y, nombres_tiendas)
plt.title('Producto más y menos vendido por tienda')
plt.legend()
plt.tight_layout()
plt.show()

print()
plt.figure(figsize=(10, 6))
plt.plot(nombres_tiendas, veces_mas_vendido, marker='o', label='Producto más vendido', color='green')
plt.plot(nombres_tiendas, veces_menos_vendido, marker='o', label='Producto menos vendido', color='red')
plt.title('Comparación de ventas por tienda')
plt.xlabel('Tiendas')
plt.ylabel('Veces vendido')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

print()
x = np.arange(len(nombres_tiendas))
width = 0.4

plt.bar(x - width/2, veces_mas_vendido, width, color='green', label='Producto más vendido')
plt.bar(x + width/2, veces_menos_vendido, width, color='red', label='Producto menos vendido')

plt.xlabel('Tiendas')
plt.ylabel('Veces vendido')
plt.title('Comparación de ventas por tienda')

plt.xticks(x, nombres_tiendas)
plt.legend()
plt.show()
```

# 5. Envío promedio por tienda

```python
#En este paso, debes calcular el costo de envío promedio para cada tienda.
#El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.
def calcular_envios(envios):
  costo_de_envio = envios['Costo de envío']
  suma_total = sum(costo_de_envio)
  cantidad_de_envio = len(costo_de_envio)
  promedio_envio = suma_total / cantidad_de_envio
  return promedio_envio
#tienda 1
promedio1 = calcular_envios(tienda)
print(f'La tienda 1 tuvo un gasto promedio de $ {promedio1:.2f} por envio')
#tienda 2
promedio2 = calcular_envios(tienda2)
print(f'La tienda 2 tuvo un gasto promedio de $ {promedio2:.2f} por envio')
#tienda 3
promedio3 = calcular_envios(tienda3)
print(f'La tienda 3 tuvo un gasto promedio de $ {promedio3:.2f} por envio')
#tienda 4
promedio4 = calcular_envios(tienda4)
print(f'La tienda 4 tuvo un gasto promedio de $ {promedio4:.2f} por envio')
```

```python
import matplotlib.pyplot as plt

nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
costo_envio_promedio = [promedio1, promedio2, promedio3, promedio4]

plt.barh(nombres_tiendas, costo_envio_promedio, color='blue')
plt.xlabel('Ingreso Total ($)')
plt.title('Gasto Promedio')
plt.tight_layout()
plt.show()

print()
plt.pie(costo_envio_promedio, labels=nombres_tiendas, autopct='%1.2f%%', startangle=90, colors=['lightcoral', 'gold', 'lightgreen', 'skyblue'])
plt.title('Distribución de Promedio Totales por Tienda')
plt.axis('equal')
plt.show()

print()
plt.plot(nombres_tiendas, costo_envio_promedio, marker='o', color='b', linestyle='-', linewidth=2, markersize=8)
plt.title('Gasto promedio por Tienda')
plt.xlabel('Tiendas')
plt.ylabel('Ingreso Total ($)')
plt.grid(True)
plt.show()
```
